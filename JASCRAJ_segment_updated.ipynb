```python
import matplotlib.pyplot as plt
import cv2

# Load your images
before = cv2.imread('/content/semantic_based_painting/images/paris2.jpeg')      # Replace with your image path
after = cv2.imread('/content/semantic_based_painting/images/out/imgs/exp_320_paris2_mask_False_mode_uniform_filter_False_brush_sizes_[0.8, 0.4, 0.2, 0.05]_budgets_[9, 49, 64, 81]_st_False_content_w_0.005/paris2_lvl_0.jpg')        # Replace with your result path

# Convert BGR (OpenCV default) to RGB for displaying correctly with matplotlib
before = cv2.cvtColor(before, cv2.COLOR_BGR2RGB)
after = cv2.cvtColor(after, cv2.COLOR_BGR2RGB)

# Plot side by side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(before)
plt.title('Before')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(after)
plt.title('After')
plt.axis('off')

plt.tight_layout()
plt.show()
```
